引入相关依赖的包

```python
from sklearn.linear_model import LinearRegression  # 导入线性回归模型
import matplotlib.pyplot as plt  # 绘图库
import numpy as np
```

构建房价数据集并可视化

```python
x = np.array([121, 125, 131, 141, 152, 161]).reshape(-1, 1)  # x是房屋面积，作为特征
y = np.array([300, 350, 425, 405, 496, 517]) # y是房屋价格，作为标签
plt.scatter(x, y)
plt.xlabel('area') # 添加横坐标：面积
plt.ylabel('price') # 添加纵坐标：价格
plt.show()
```

模型训练

```python
lr = LinearRegression()  # 将线性回归模型封装为对象
lr.fit(x, y)  # 模型在数据上训练
```

模型的可视化

```python
w = lr.coef_[0]  # 模型的斜率
b = lr.intercept_  # 模型的截距
print('斜率：', w)
print('截距：', b)
```

```python
plt.scatter(x, y)
plt.xlabel('area') # 添加横坐标：面积
plt.ylabel('price') # 添加纵坐标：价格
plt.plot([x[0], x[-1]], [w * x[0] + b, w * x[-1] + b])
```

模型预测

```python
testX = np.array([[130]]) # 测试样本，面积为130
lr.predict(testX)
```
